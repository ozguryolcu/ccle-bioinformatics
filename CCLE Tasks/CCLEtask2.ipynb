```python
import os 
import pandas as pd
import numpy as np

os.mkdir(r"C:\\Users\\ozgur\\Desktop\\CCLE")
```

```python
df1 = pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE_NP24.2009_Drug_data_2015.02.24 (2).csv")
dbDrug = pd.DataFrame(df1)
```

```python
dbDrug.head()
```

```python
def getCancerType(cellLine=str):
        x = cellLine.split("_")
        y = " ".join(x[1:])
        if y == "":
                return "UNDEFINED"
        return y

dbDrug["Cancer Type"] = dbDrug["CCLE Cell Line Name"].apply(getCancerType)
dbDrug["Cancer Type"].value_counts()
```

```python
for x in dbDrug["Cancer Type"].unique():
    
    if x == "UNDEFINED":
        continue
    else:
        os.mkdir("C:\\Users\\ozgur\\Desktop\\CCLE\\{}".format(x))
    
    for y in dbDrug["Compound"].unique():
        temp = dbDrug.loc[(dbDrug["Cancer Type"] == x) & (dbDrug["Compound"] == y)]
        temp[["Cancer Type", "Primary Cell Line Name", "Compound", "IC50 (uM)", "ActArea"]].to_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\{}\\{}".format(x,y), index = False)
```
