```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
from scipy.stats import pearsonr as pr
from scipy.stats import spearmanr as spr

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df1 = pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CCLE_NP24.2009_Drug_data_2015.02.24 (2).csv")
df2 = pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CCLE_RNAseq.txt", sep="\t")
dbDrug = pd.DataFrame(df1)
dbBreastExp = pd.DataFrame(df2)
```

```python
dbBreastExp.head()
```

```python
def getCancerType(cellLine=str):
        x = cellLine.split("_")
        y = " ".join(x[1:])
        if y == "":
                return "UNDEFINED"
        return y

dbDrug["Cancer Type"] = dbDrug["CCLE Cell Line Name"].apply(getCancerType)
dbDrug.head()
```

```python
temp = dbDrug.loc[dbDrug["Cancer Type"] == "BREAST"]
dbBreastSubSet = temp[["Cancer Type", "CCLE Cell Line Name", "Compound", "IC50 (uM)", "ActArea"]]
```

```python
dbBreastSubSet.head()
```

```python
listOfCellLines = dbBreastExp.columns[2:]
print(listOfCellLines)
```

```python
temp = dbBreastSubSet["CCLE Cell Line Name"].isin(listOfCellLines)
dbBreastSubSet = dbBreastSubSet[temp]
```

```python
uniqueList = dbBreastSubSet["CCLE Cell Line Name"].unique()
for x in dbBreastExp.columns:
    if x == "Hugo_Symbol" or x == "Ensembl_Gene_Id":
        continue
    elif x not in uniqueList:
        dbBreastExp.drop(columns=x, inplace= True)
```

```python
dbBreastExp.head()
```

```python
os.mkdir(r"C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson")
os.mkdir(r"C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\IC50 VALUES")
tempList = dbBreastSubSet["Compound"].unique()
columns = dbBreastExp.columns[2:]

for x in tempList:
    subset = dbBreastSubSet.loc[dbBreastSubSet["Compound"] == x]
    val = 0
    geneArray = []
    pearsonArray = []
    pValueArray = []
    compoundArray = []

    try:
        for t in dbBreastExp["Hugo_Symbol"]:
            if len(columns) == len(subset["CCLE Cell Line Name"]):
                if np.std(dbBreastExp.iloc[val,2:].astype(float)) == 0 or np.std(subset["IC50 (uM)"].astype(float)) == 0:
                    val+=1
                    continue
                else:
                    corr_coef, p_value = pr(dbBreastExp.iloc[val,2:].astype(float), subset["IC50 (uM)"].astype(float))
                    pearsonArray.append(corr_coef)
                    geneArray.append(dbBreastExp.iloc[val, 1])
                    pValueArray.append(p_value)
                    compoundArray.append(x)
                    val += 1
            else:
                temp5 = list(columns.isin(subset["CCLE Cell Line Name"]))
                temp5.insert(0, True)
                temp5.insert(0, True)
                dbBreastExp2 = dbBreastExp.loc[:, temp5]
                if np.std(dbBreastExp2.iloc[val,2:].astype(float)) == 0 or np.std(subset["IC50 (uM)"].astype(float)) == 0:
                    val += 1
                    continue
                else: 
                    corr_coef, p_value = pr(dbBreastExp2.iloc[val,2:].astype(float), subset["IC50 (uM)"].astype(float))
                    pearsonArray.append(corr_coef)
                    geneArray.append(dbBreastExp.iloc[val, 1])
                    pValueArray.append(p_value)
                    compoundArray.append(x)
                    val += 1
    except IndexError:
        pass
    finally:
        temp4 = pd.DataFrame(list(zip(geneArray, compoundArray, pearsonArray, pValueArray)), columns= ["Gene", "Compound", "IC50 PearsonCorrelation", "P-Values"])
        temp4.sort_values(by=["P-Values"]).to_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\IC50 VALUES\\{} IC50 PearsonCorr".format(x), index= False, sep="\t") 
```
