```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

pd.options.display.float_format = "{:,.12f}".format
pd.options.display.max_columns = 20
pd.options.display.max_rows = 20
```

```python
placeholder = []
for x in os.listdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\IC50 VALUES"):
    placeholder.append(pd.DataFrame(pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\IC50 VALUES\\{}".format(x), sep="\t")))
pearsonIC50 = pd.concat(placeholder, ignore_index=True)

placeholder2 = []
for y in os.listdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\ActArea VALUES"):
    placeholder2.append(pd.DataFrame(pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Pearson\\ActArea VALUES\\{}".format(y), sep="\t")))
pearsonActArea = pd.concat(placeholder2, ignore_index=True)

placeholder3 = []
for z in os.listdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Spearman\\IC50 VALUES"):
    placeholder3.append(pd.DataFrame(pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Spearman\\IC50 VALUES\\{}".format(z), sep="\t")))
spearmanIC50 = pd.concat(placeholder3, ignore_index=True)

placeholder4 = []
for t in os.listdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Spearman\\ActArea VALUES"):
    placeholder4.append(pd.DataFrame(pd.read_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Spearman\\ActArea VALUES\\{}".format(t), sep="\t")))
spearmanActArea = pd.concat(placeholder4, ignore_index=True)
```

```python
os.mkdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Filtered")
os.mkdir("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Filtered\\IC50-ActArea-Pearson")
subset = pearsonIC50.loc[pearsonIC50["P-Values"] < 0.05]
subset2 = pearsonActArea.loc[pearsonActArea["P-Values"] < 0.05]

for x in subset["Compound"].unique():
    placeholder5 = subset.loc[subset["Compound"] == x]
    placeholder6 = subset2.loc[subset2["Compound"] == x]

    pearsonFiltered = pd.merge(placeholder5, placeholder6, on=["Gene", "Compound"], how="inner").reset_index()
    pearsonFiltered.drop(["index"], axis=1).to_csv("C:\\Users\\ozgur\\Desktop\\CCLE\\CorrelationValues\\Filtered\\IC50-ActArea-Pearson\\{}".format(x), index= False, sep="\t")
```
